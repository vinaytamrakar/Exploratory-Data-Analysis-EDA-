```python
Ashneer_present=len(shktnk[shktnk.AshneerGrover_present==1])
Anupam_present=len(shktnk[shktnk.AnupamMittal_present==1])
Aman_present=len(shktnk[shktnk.AmanGupta_present==1])
Namita_present=len(shktnk[shktnk.NamitaThapar_present==1])
Vinneta_present=len(shktnk[shktnk.VineetaSingh_present==1])
Peyush_present=len(shktnk[shktnk.PeyushBansal_present==1])
Ghazal_present=len(shktnk[shktnk.GhazalAlagh_present==1])
```

```python
fig=plt.figure(figsize=(10,5))
present=[Ashneer_present,Anupam_present,Aman_present,Namita_present,Vinneta_present,Peyush_present,Ghazal_present]
name=['Ashneer','Anupam','Aman','Namita','Vinneta','Peyush','Ghazal']
plt.bar(name,present,color='maroon',width=0.4)
plt.xlabel("Candidate Name")
plt.ylabel("Candidate Present")
plt.title("Candidate Present vs Name")
plt.show()
```

```python
shktnk['Deal '].value_counts()
```

```python
# PIE chart on Percentage of candidates deal

plt.pie(present,labels=name,radius=1.5,autopct='%0.1f%%')
plt.suptitle('Percentage of candidates deal',x=0.55,y=1.15)
plt.show()
```
